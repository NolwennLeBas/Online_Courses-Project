```python
from time import sleep

subject = []
title=[]
rating=[]
number_of_ratings=[]
number_of_students=[]
price=[]

for url in df['course_url']:
    response = requests.get(url)
    if response.status_code==200:
        soup = BeautifulSoup(response.content, "html.parser")
        
        try:
            subject_1 = soup.select("ul > li:nth-child(2) > a")[2].get_text()
            subject.append(subject_1)
        except:
            subject.append("None")
        
        try:
            title_1=soup.select("div.col-md-7.pr-4 > h1")[0].get_text()
            title.append(title_1)
        except:
            title.append("None")
        
        try:
            rating_1 = soup.select("div.h5.ml-1.mr-3.mb-0")[0].get_text()
            rating.append(rating_1.split()[0])
        except:
            rating.append("None")
        
        try:
            number_of_ratings_1=soup.select("div.micro")[0].get_text()
            number_of_ratings_1=number_of_ratings_1.split()[0].replace(',','')
            number_of_ratings.append(number_of_ratings_1)
        except:
            number_of_ratings.append("None")
        
        try:
            number_of_students_1 = soup.select("div:nth-child(1) > div.small > div")[0].get_text()
            number_of_students_1 = number_of_students_1.split()[0].replace(',','')
            number_of_students.append(number_of_students_1)
        except:
            number_of_students.append("Unknown")
        
        try:
            price_1 = soup.select("td:nth-child(2) > p")[0].get_text()
            price_1=price_1.split()[0]
            price_1=price_1.replace('$','')
            price.append(price_1)
        except:
            price.append("Unknown")
        
        sleep(2)
        
    elif response.status_code!=200:
        subject.append("None")
        title.append("None")
        rating.append("None")
        number_of_ratings.append("None")
        number_of_students.append("None")
        price.append("None")

        
print(len(subject))
print(len(title))
print(len(rating))
print(len(number_of_ratings))
print(len(number_of_students))
print(len(price))
```

```python
df['subject']=subject
df['title']=title
df['rating']=rating
df['number_of_ratings']=number_of_ratings
df['number_of_students']=number_of_students
df['price']=price
df
```

```python
df.to_csv('updated_edx_courses.csv', index=False)
```
