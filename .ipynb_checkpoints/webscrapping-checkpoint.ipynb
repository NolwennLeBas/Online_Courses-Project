```python
from time import sleep

subject = []
title=[]
rating=[]
number_of_ratings=[]
update_date=[]
number_of_students=[]

for url in df_study['url']:
    response = requests.get(url)
    if response.status_code==200:
        soup = BeautifulSoup(response.content, "html.parser")
        
        try:
            subject_1 = soup.select(".ud-heading-sm")[3].get_text()
            subject.append(subject_1)
        except:
            subject.append("None")
        
        try:
            title_1=soup.select(".ud-heading-xl")[0].get_text()
            title.append(title_1)
        except:
            title.append("None")
        
        try:
            rating_1 = soup.select("span.ud-sr-only")[1].get_text()
            rating.append(rating_1.split()[1])
        except:
            rating.append("None")
        
        try:
            number_of_ratings_1=soup.select("span:nth-child(2)")[6].get_text()
            number_of_ratings_1=number_of_ratings_1.split()[0]
            number_of_ratings.append(number_of_ratings_1[1:])
        except:
            number_of_ratings.append("None")
        
        try:
            update_date_1=soup.select("span:nth-child(2)")[7].get_text()
            update_date.append(update_date_1.split()[-1])
        except:
            update_date.append("None")
        
        try:
            number_of_students_1 = soup.select(".enrollment")[0].get_text()
            number_of_students_1 = number_of_students_1.split()[0].replace(',','')
            number_of_students.append(number_of_students_1)
        except:
            number_of_students.append("None")
        
        sleep(2)
        
    elif response.status_code!=200:
        subject.append("None")
        title.append("None")
        rating.append("None")
        number_of_ratings.append("None")
        update_date.append("None")
        number_of_students.append("None")

        
print(len(subject))
print(len(title))
print(len(rating))
print(len(number_of_ratings))
print(len(update_date))
print(len(number_of_students))
    
```

```python
df_study=df_study.reset_index(drop=True)
df_study
```

```python
df_study['subject']=subject
df_study
```

```python
df_study['course_title']=title
df_study
```

```python
df_study['rating']=rating
df_study
```

```python
df_study['number_of_ratings']=number_of_ratings
df_study['update_date']=update_date
df_study['number_of_students']=number_of_students
df_study
```

```python
df_study.to_csv('updated_udemy_courses.csv', index=False)
```
